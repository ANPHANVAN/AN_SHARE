# *Machine Learning*

machine learning to predict total sales forecasting

## Spliting data

````python
df.columns
````

````python
# Split data into features and labels
features = df[['PROD_QTY', 'PACK_SIZE','BRAND', 'LIFESTAGE',
       'PREMIUM_CUSTOMER']]
labels = df['TOT_SALES']
````

````python
# Split into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(features, labels, test_size=0.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, random_state=1)
````

## Data type classification

````python
# Specify which columns are numerical and which are categorical
numeric_features = ['PROD_QTY']
categorical_features = ['PACK_SIZE','BRAND', 'LIFESTAGE','PREMIUM_CUSTOMER']

# Create a ColumnTransformer to apply different transformations
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),    # Scale numeric features
        ('cat', OneHotEncoder(), categorical_features)  # Encode categorical features
    ])
````

## Create model

````python
# Create a pipeline that first preprocesses the data, then fits the model
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', DecisionTreeRegressor())
])
# Fit the model
model.fit(X_train, y_train)
````

## Test score model with data train , data validation and data final test

````python
train_r2_score = model. score (X_train, y_train) 
test_r2_score = model.score (X_val, y_val) 
print (f"R2 score of train set : {train_r2_score}") 
print(f"R2 score of valid set : {test_r2_score}")
````

````python
train_r2_score = model.score (X_train, y_train) 
test_r2_score = model.score (X_test, y_test) 
print (f"R2 score of train set : {train_r2_score}") 
print(f"R2 score of test set : {test_r2_score}")
````

> Mô hình có độ chính xác tốt với tập dữ liệu test, sự chính xác tren 99%

## See a detail predictions

````python
# Make predictions
predictions = model.predict(X_test)

# Thêm cột predictions vào DataFrame
X_test_predictions = X_test
X_test_predictions['Predictions'] = predictions

# Hiển thị bảng kết quả
X_test_predictions
````
